```python
import numpy as np
import pandas as pd
from apyori import apriori
```

```python
market_df = pd.read_csv('Market Basket Analysis.csv')
```

```python
market_df.head()
```

```python
#create a list of lists with the input dataframe
market_list = []

for i in range(5):
    market_list.append([str(market_df.values[i,j]) for j in range(5)])
```

```python
market_list
```

```python
rules = apriori(market_list, min_support = 0.2, min_confidence = 1, min_lift = 5, min_length=3)
```

```python
result = list(rules)
```
