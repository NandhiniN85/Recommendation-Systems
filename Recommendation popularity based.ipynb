```python
import pandas as pd
import copy as copy
```

```python
movies_df = pd.read_csv('movies.csv')

movies_copy_df = copy.deepcopy(movies_df)

movies_copy_df.head()
```

```python
movies_copy_df.shape
```

```python
ratings_df = pd.read_csv('ratings.csv')

ratings_copy_df = copy.deepcopy(ratings_df)

ratings_copy_df.head()
```

```python
ratings_copy_df.shape
```

```python
movies_ratings_df = pd.merge(movies_df, ratings_df, on='movieId')
```

```python
movies_ratings_df
```

```python
ratings_mean_count = pd.DataFrame(movies_ratings_df.groupby('title')['rating'].mean()) 
```

```python
ratings_mean_count['count'] = pd.DataFrame(movies_ratings_df.groupby('title')['rating'].count())
```

```python
ratings_mean_count.sort_values('rating', ascending=False)
```

Using Mean and count values, the popularity of the movie is determined.
