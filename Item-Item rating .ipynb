```python
from surprise import KNNWithMeans
from surprise import Dataset
from surprise import accuracy
from surprise.model_selection import train_test_split
from surprise import Reader
import pandas as pd
```

```python
ratings_df = pd.read_csv('ratings.csv').drop('timestamp', axis=1)
```

```python
ratings_df
```

```python
reader = Reader()
```

```python
data = Dataset.load_from_df(ratings_df, reader)
```

```python
trainset, testset = train_test_split(data, test_size=0.15)
```

```python
knnwithmeans =  KNNWithMeans(k=50, sim_options = {'name':'pearson_baseline', 'user_based':False})
knnwithmeans.fit(trainset)
```

```python
predict = knnwithmeans.test(testset)
```

```python
predict
```

```python
accuracy.mse(predict)
```
